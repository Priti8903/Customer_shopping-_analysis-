```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={'purchase_amount_(usd)':"purchase_amount"})
```

```python
df.columns
```

```python
# create  acolumn name age_group
df["age_group"]=pd.qcut(df["age"],4,labels=["Young","Adult","Middle-Age","Senior"])
```

```python
df[["age","age_group"]].head(10)
```

```python
#create a column purchese_frequency_days
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Bi-Weekly":14,
    "Quarterly":90,
    "Annually":365,
    "Every 3 months":90
}
df["purchese_frequency_days"]=df['frequency_of_purchases'].map(frequency_mapping)
```
